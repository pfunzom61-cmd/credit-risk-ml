# Data Ingestion - Home Credit Default Risk

Goal: Load raw CSV data and insert into PostgreSQL database (credit_risk).


Load data from raw applicaton csv to gain a basic overview of the dataset

```python
import pandas as pd
from sqlalchemy import create_engine

#Load a specific set of columns (first batch)
columns_to_import = [
    'SK_ID_CURR',
    'TARGET',
    'AMT_INCOME_TOTAL',
    'NAME_CONTRACT_TYPE',
    'CODE_GENDER'
]

# 4️ Connect to PostgreSQL
engine = create_engine('postgresql+psycopg2://postgres:PfunzoPostgres@localhost:5432/credit_risk')

df = pd.read_csv("../Data/raw/application_train.csv", usecols=columns_to_import)

print(df.head())
print(df.shape)
print(df.info())
```

Covert data types to match schema data types and validate if all primary keys are unique before importing

```python
# Convert SK_ID_CURR → BIGINT equivalent
df["SK_ID_CURR"] = df["SK_ID_CURR"].astype("int64")

# Convert TARGET → SMALLINT equivalent
df["TARGET"] = df["TARGET"].astype("int8")

# Convert AMT_INCOME_TOTAL → NUMERIC(12,2) equivalent
df["AMT_INCOME_TOTAL"] = df["AMT_INCOME_TOTAL"].round(2)

# Convert categorical columns → string type
df["NAME_CONTRACT_TYPE"] = df["NAME_CONTRACT_TYPE"].astype("string")
df["CODE_GENDER"] = df["CODE_GENDER"].astype("string")

# Display the new types
print(df.dtypes)

df["SK_ID_CURR"].is_unique #Checks if all primary keys are unique to avoid duplicates in the database
```

First inserion into database using SQLAlechmy

```python
#Insert first 5 columns of data into the database

df.columns = df.columns.str.lower() # Convert column names to lowercase to match the database schema
df.to_sql(
    "application_train",
    engine,
    if_exists="append",
    index=False
)

```

Create the new application_train schema and copy output in Dbeaver. This ensures that all 122 coloumns are created in the data base so we can begin the data alignment process

```python
def generate_sql_schema(df, table_name):
    sql_query = f"CREATE TABLE {table_name} (\n"
    columns = []
    
    for col, dtype in df.dtypes.items():
        # Mapping Logic
        if col == 'SK_ID_CURR':
            definition = f"    {col.lower()} BIGINT PRIMARY KEY"
        elif col == 'TARGET':
            definition = f"    {col.lower()} SMALLINT NOT NULL"
        elif dtype == 'int64':
            definition = f"    {col.lower()} BIGINT"
        elif dtype == 'float64':
            # 14 total digits, 3 after the decimal
            definition = f"    {col.lower()} NUMERIC(14,3)"
        else:
            # For objects/strings
            definition = f"    {col.lower()} VARCHAR(100)"
            
        columns.append(definition)
    
    sql_query += ",\n".join(columns)
    sql_query += "\n);"
    return sql_query

# Run it
schema_sql = generate_sql_schema(df, "application_train")
print(schema_sql)
```
